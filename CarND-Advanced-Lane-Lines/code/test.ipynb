```python
import cv2
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

% matplotlib inline

image = cv2.imread('./frames/picture-41.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# cv2.rectangle(image,(65, 640), (365, 720), (0,255,0), 2)
# cv2.rectangle(image,(500, 470), (1100, 720), (0,255,0), 2)
#     src = np.float32([ [500, 470],[950, 470], [1100, 720],  [180, 720]])

cv2.fillPoly(image, pts=np.asarray([[[550, 50], [800, 450], [1150, 720],  [180, 720]]], 'int32'), color=(255,0,0))


hls = cv2.cvtColor(image, cv2.COLOR_RGB2HLS).astype(np.float)

h = hls[:,:,0]

h_bin = np.zeros_like(h)
h_bin[(h>=19) & (h<=21)] = 1

color = np.uint8(np.dstack((h_bin, np.zeros_like(h),  np.zeros_like(h)))) * 255


plt.figure()
# plt.imshow(image)

result = cv2.addWeighted(image, 1, color, 0.3, 0)
plt.imshow(result)
```
